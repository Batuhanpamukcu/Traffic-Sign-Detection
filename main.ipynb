## Eğitim Görüntüsü - 1

```python
# Kütüphaneler tanımlandı
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

%matplotlib inline

# Görüntü - 1 okundu
original_image=cv.imread("egitim_goruntu1.pgm") #görüntü okundu

# Görüntü - 1 çizdirildi
plt.figure()
plt.imshow(original_image, cmap="gray") # cmap="gray" demezsek renkli
plt.title("Eğitim Görüntüsü - 1 Orijinal Görsel")
plt.colorbar()

# Orijinal görsele median filtresi uygulandı (filtreleme)
medianFiltered_image = cv.medianBlur(original_image,5)

# Cannny algoritması ile median uygulanmış görseldeki sınırlar tespit edildi (ayrıt saptama)
# Burada seçilen lower ve upper threshold değerleri görseller için en optimum değerler olacak şekilde seçilmiştir. (390-420)
edgeDetected_image = cv.Canny(medianFiltered_image,390,420, L2gradient = True)

# Filtrelenip sınrıları tespit edilen görüntü çizdirildi
plt.figure()
plt.imshow(edgeDetected_image, cmap="gray") # cmap="gray" demezsek renkli
plt.title("Eğitim Görüntüsü - 1 Filtreleme ve Canny Ayrıt Saptama")
plt.colorbar()

#Global thresholding ayrıt saptama sonrası uygulandı (eşikleme)
(thresh, thresholded_image) = cv.threshold(edgeDetected_image, 200, 255, cv.THRESH_BINARY)

# morfolojik operasyonlar için maske tanımlandı
kernel = np.ones((5,5),np.uint8)

# dilation işlemi tanımlanan maske ile uygulandı
dilation_image = cv.dilate(thresholded_image,kernel,iterations = 1)

#dilation uygulanmış görüntü çizdirildi
plt.figure()
plt.imshow(dilation_image, cmap="gray") # cmap="gray" demezsek renkli
plt.title("Eğitim Görüntüsü - 1 Dilation")
plt.colorbar()

# erosion işlemi tanımlanan maske ile uygulandı
erosion_image = cv.erode(dilation_image,kernel,iterations = 1)

#erosion uygulanmış görüntü çizdirildi
plt.figure()
plt.imshow(erosion_image, cmap="gray") # cmap="gray" demezsek renkli
plt.title("Eğitim Görüntüsü - 1 Erosion")
plt.colorbar()

# görüntüde kalan gürültünün daha rahat incelenebilmesi için negatifi alındı ve çizdirildi
negative_image = cv.bitwise_not(erosion_image)

plt.figure()
plt.imshow(negative_image, cmap="gray") # cmap="gray" demezsek renkli
plt.title("Eğitim Görüntüsü - 1 Negative")
plt.colorbar()

# Probabilistic Hough Line Transform ile çizgi tespiti yapıldı 
lineDetected_image = cv.HoughLinesP(dilation_image, 1, np.pi/180,30) # tespit için dilate edilmiş görüntü kullanıldı
for line in lineDetected_image:
    x1,y1,x2,y2 = line[0]
    cv.line(original_image,(x1,y1),(x2,y2),(0,255,0),4) # tespit edilen tabela çizgileri orijinal görüntü üzerinde belirtilerek tabela tespiti gerçekleştirildi

# Sonuç görseli çizdirildi
plt.figure()
plt.imshow(original_image, cmap="gray") # cmap="gray" demezsek renkli
plt.title("Eğitim Görüntüsü - 1 Tabela Tespiti")
plt.colorbar()
```
